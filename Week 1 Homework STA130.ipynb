```python
#Task 7

"""
1.
df.dropna() is a command to remove all rows or columns that contain a null data entry. 
- You may chose to use to use df.dropna() over del df['column'] if you are trying to clear all missing entries in a 
  large data frame.

2.
del df['column'] is a command to remove a specified column from the data frame.
- You may chose to use del df['column'] over df.dropna() when you know which column to remove as the user has control 
  over the scope of the function


3. If you know ahead of time that columns are not useful, you should use del df['column'] to delete them first.
This helps avoid the accidental deletion of entries when using df.dropna(), as the del df['column'] with null values has 
been removed.

"""

#4. Before Removal:
df_copy = df.copy()
print(df_copy.shape)
print("")
print(df_copy.isna().sum())
print("")
print(df_copy.describe())

#REMOVAL 
del df_copy['age']
del df_copy['deck']
df_copy.dropna()
```

```python
#AFTER

print(df_copy.shape)
print("")
print(df_copy.isna().sum())
print("")
print(df_copy.describe())
```
